```python
#Copyright 2020 Awase Khirni Syed 
#University of Waterloo, Geospatial Sensing and Data Intelligence Lab,
# feature extraction from topographic maps to render holographic/mixed reality perspectives
#https://uwaterloo.ca/geospatial-sensing/people-profiles/category/45
import pandas as pd 
import numpy as np 
from skimage.io import imread, imshow 
#skeletonize library
#it reduces binary objects to 1 pixel wide representation. the can be useful for feature extraction, and/ or representing an objects topology
from skimage.morphology import skeletonize 
from skimage import data 
import matplotlib.pyplot as plt 
from skimage.util import invert 


#reading the image from url 
image = imread('https://prd-wret.s3.us-west-2.amazonaws.com/assets/palladium/production/s3fs-public/thumbnails/image/_352890_BrowseImage_0.jpg')
imshow(image)
#checking image shape 
image.shape, imshow(image)


#skeletonize process 
map_skeleton = skeletonize(image)
 
# display results
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 4),
                         sharex=True, sharey=True)

ax = axes.ravel()

ax[0].imshow(image, cmap=plt.cm.gray)
ax[0].axis('off')
ax[0].set_title('original', fontsize=20)

ax[1].imshow(map_skeleton, cmap=plt.cm.gray)
ax[1].axis('off')
ax[1].set_title('skeleton', fontsize=20)

fig.tight_layout()
plt.show()
```

```python
#experiment 2 with inverse color image 
#applying zhangs method => a fast parallel for thining digital pattersns. ACM March 1984, vol27, no.3
#reading the image from url 
inv_image = imread('F:/topomapdata/_352890_InvertedImage.jpg')
imshow(inv_image)
#checking image shape 
image.shape, imshow(inv_image)


#skeletonize process 
inv_map_skeleton = skeletonize(inv_image)
 
# display results
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 4),
                         sharex=True, sharey=True)

ax = axes.ravel()

ax[0].imshow(inv_image, cmap=plt.cm.white)
ax[0].axis('off')
ax[0].set_title('inversion_original', fontsize=20)


ax[1].imshow(inv_map_skeleton, cmap=plt.cm.white)
ax[1].axis('off')
ax[1].set_title('skeleton', fontsize=20)

fig.tight_layout()
plt.show()
```
